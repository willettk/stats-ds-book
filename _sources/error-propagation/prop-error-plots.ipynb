```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
#For 3d plots
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.random.normal(1,.2,1000)
y = np.random.normal(2,.4,1000)
```

```python
z = 1./x
```

## With Matplotlib

```python
plt.scatter(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, alpha=0.3)
```

```python
plt.scatter(x,z)
```

```python
plt.scatter(y,z)
```

## With Seaborn

```python
import seaborn as sns
```

```python
sns.distplot(x)
```

```python
g = sns.JointGrid(x,z)
g.set_axis_labels('x','x/y')
g = g.plot_joint(plt.scatter, alpha=0.5)
g = g.plot_marginals(sns.distplot)
```

```python
#with density plots
g = sns.JointGrid(x,z)
g.set_axis_labels('x','x/y')
g = g.plot_joint(sns.kdeplot, cmap="Blues_d")
g = g.plot_marginals(sns.kdeplot, shade=True)
```

```python
#with regression curve
g = sns.JointGrid(x,z)
g.set_axis_labels('x','z')
g = g.plot(sns.regplot, sns.distplot)
```
